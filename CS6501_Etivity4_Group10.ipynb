# **Introduction**

##**Genetic Algorithms**

“A genetic algorithm (GA) is a heuristic search algorithm used to solve search and optimisation problems. This algorithm is a subset of evolutionary algorithms, which are used in computation” (Muthee, 2021). The behaviour of chromosomes and their structure impact the GA, possible solutions are provided by every chromosome. There is also a fitness function, explained in greater depth below, and the quality of this function is directly linked to how good the solution is. GAs can be applied in numerous fields in the real-world such as transport, DNA analysis, economics, and aircraft design. 

Like any model/algorithm there are limitations to the GA. The GA is not suitable for solving problems that are relatively simple, there is no guarantee that a final solution will be of a good enough quality and repeating the fitness value calculations may result in some computational problems. 


##**Greedy Algorithms**

A greedy algorithm is an algorithm that can be used for optimisation problems. The algorithm seeks to find the best way to solve the problem in its entirety and in so doing seeks to make the best choice at every step in solving the problem. Greedy algorithms are useful and successful in certain problems such as the travelling salesman problem, problems in which the shortest past through a list/graph of points needs to be found. As useful as they are, greedy algorithms may not always produce the optimal solution. In such a scenario, the failure may be down to the fact that the algorithm did not consider al the data. 

The greedy nearest neighbour algorithm can be used to solve the travelling salesman problem and is quite accurate in its approximation. This algorithm can be executed quickly but can sometimes miss shorter/more optimal routes due to its “greedy” nature. 


##**One-Max Problem**

In the one max problem evolution is used to find a specific “gene”. In a machine learning context, a gene is a piece of text/binary string that is filled with 0’s and 1’s. The aim of the one max problem is to find a binary string that matches a specified target. To do this, we must use an evolutionary algorithm which will evolve our population until our target is realised and emerges (github.com, n.d.)

In the problem, a population of individuals is created that consists of integer vectors that are filled with 0 and 1 randomly. Then the evolutionary algorithm allows the population to evolve until no 0 values are seen in one of the members of the population, only 1s. To solve the one max problem, we need to generate a population. Throughout solving the problem, we will add some functions and operators that take care of the evaluation and the evolution of our population (deap.readthedocs.io, n.d.)

The one max problem is a trivial problem and is often used for introducing the topic of genetic algorithms. Despite its simplicity, it gives valuable insights into many critical aspects of genetic algorithms (Moriarity, 2021).
As mentioned previously, the one max problem is great for gaining an insight into the topic of genetic algorithms and some of the critical aspects of them. “Genetic algorithms are a class of optimisation algorithms based on evolution and natural selection. They use strategies loosely based on genetics and biology to produce optimal – think “best” – or near-optimal solutions to complicated problems” (Moriarity, 2021). Initially, these genetic algorithms’ intended use was to be a technique that created adaptable programs. Nowadays, genetic algorithms have many applications and are used in fields such as artificial intelligence and finance. 


##**One-Max - Fitness Function**

In genetic algorithms, there is always a fitness function. A fitness function is a function that is used to evaluate “how close a given solution is to the optimum solution of the desired problem. It determines how fit a solution is” (Mallawaarachchi, 2017). 

A fitness function is used because within genetic algorithms, each solution is a string of binary numbers. These strings are known as chromosomes. The solutions need to be tested so that we can come up with the best set of solutions to a certain problem. To determine how close a solution was to meeting a specified target, each solution is scored. Applying the fitness function to the results of the test is what generates the score. 

There are certain requirements that must be met for any fitness function. The function should be clearly defined, it should be implemented in an efficient manner it should measure how accurate a given solution is in solving the problem at hand and it should generate intuitive results. 
Ultimately, a fitness function that gives low values for poor solutions and higher values for good solutions is desired.


##**Travelling Salesman Problem** 

The travelling salesman problem is concerned with finding “the shortest possible route that visits every city exactly once and returns to the starting point” (geeksforgeeks.org, 2018) given a set of cities and the distances between every pair of them. 

The goal/problem to be solved is finding the shortest path that a salesman should take through a set of cities and return to the origin city. In the problem, both the list of cities and the distance between each pair is provided. There are 3 methods that can be used to solve this problem, dynamic programming, simulated annealing, and 2-opt. all of which will be discussed briefly. 

**Dynamic Programming:** this method always finds the optimum answer to the travelling salesman problem. A drawback to this method is that the “time complexity would exponentially increase with the number of cities” (Ataee, 2020). This method needs more time as the number of cities go up and despite guaranteeing the optimum solution, the time and computational power needed often cannot be afforded in a real-life scenario.

**Simulated Annealing:** simulated annealing (SA) is a heuristic algorithm and there are two major rules within it. 
1.	In order to avoid being trapped in a local optimum, the moving direction must be determined in each step of the process.
2.	Once the process is getting close to the end, the size of the search step must be reduced. This allows quicker and more rigorous movement during the early steps while also providing for more cautious movement in later steps. 
Within the SA method, the search process continues until either a satisfactory solution is found or until the criteria for stopping have been met. One drawback of this method is that is that it is insensitive to the formulation and tuning of its parameters (search step and moving direction).

**2-opt:** “The main idea behind the 2-opt method is to remove path crossing in each neighbourhood of cities” (Ataee, 2020). In this method the problem is considered solved when a good-enough solution has been converged by the algorithm, even if it is sub-optimal. Compared to the previous two methods of solving the travelling salesman problem, this method is much faster and easier to implement.

#Task 1 - OneMax Problem

## Parameters

```python
## Problem parameters:
# Length of the bit string to be optimized
# each bit is considered as a gen in the encoding
ONE_MAX_LENGTH = 10  

## Genetic Algorithm parameters:
POPULATION_SIZE = 2
# probability for crossover
P_CROSSOVER = 0.7  
# probability for mutating an individual
P_MUTATION = 0.01  
# Number of generations (iterations)
MAX_GENERATIONS = 50
# Number of best solutions to save
HALL_OF_FAME_SIZE = 10
# Random seed
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
```

## Population

```python
toolbox = base.Toolbox()

# create an operator that randomly returns 0 or 1:
toolbox.register("zeroOrOne", random.randint, 0, 1)

# define a single objective, maximizing fitness strategy:
creator.create("FitnessMax", base.Fitness, weights=(1.0,))

# create the Individual class based on list:
creator.create("Individual", list, fitness=creator.FitnessMax)

# create the individual operator to fill up an Individual instance:
toolbox.register("individualCreator", tools.initRepeat, creator.Individual, toolbox.zeroOrOne, ONE_MAX_LENGTH)

# create the population operator to generate a list of individuals:
toolbox.register("populationCreator", tools.initRepeat, list, toolbox.individualCreator)
```

## Evaluation

```python
# Fitness function
# compute the number of '1's in the individual
# each bit position 
def oneMaxFitness(individual):
    # return a tuple
    return sum(individual),  

toolbox.register("evaluate", oneMaxFitness)
```

## Genetic Operations

```python
# genetic operators:

# Roulette selection:
toolbox.register("select", tools.selRoulette)

# Single-point crossover:
toolbox.register("mate", tools.cxOnePoint)

# Flip-bit mutation:
# indpb: Independent probability for each attribute to be flipped
toolbox.register("mutate", tools.mutFlipBit, indpb=1.0/ONE_MAX_LENGTH)
```

## Genetic Algorithm

```python
# Genetic Algorithm :
def main(pop_size, printresults=True, question=0):
    # Random Seed
    random.seed(RANDOM_SEED)

    # create initial population (generation 0):
    population = toolbox.populationCreator(n=pop_size)

    # define the hall-of-fame object:
    hof = tools.HallOfFame(HALL_OF_FAME_SIZE)

    # prepare the statistics object:
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("max", np.max)
    stats.register("avg", np.mean)

    # perform the Genetic Algorithm :
    population, logbook = algorithms.eaSimple(population, toolbox, cxpb=P_CROSSOVER, mutpb=P_MUTATION,
                                              ngen=MAX_GENERATIONS,
                                              stats=stats, halloffame=hof, verbose=False)

    # Genetic Algorithm is done - extract max and average statistics:
    maxFitnessValues, meanFitnessValues = logbook.select("max", "avg")

    if printresults == True:
            # print Hall of Fame:
        print("Hall of Fame Individuals = ", *hof.items, sep="\n")
        print("Best Ever Individual = ", hof.items[0])
        
        # Genetic Algorithm is done - plot max and average statistics:
        #sns.set_style("whitegrid")
        plt.plot(maxFitnessValues, color='red',label='Maximum')
        plt.plot(meanFitnessValues, color='green', label='Average')
        plt.xlabel('Generation')
        plt.ylabel('Max / Average Fitness')
        plt.title('Max and Average Fitness over Generations')
        plt.legend(loc='lower right')
        plt.show()
    
    if question == 1:
        if np.max(maxFitnessValues) == 10:
            return(True, pop_size, maxFitnessValues.index(10))
        else:
            return(False, 0,0)
```

```python
# Changing Toolbox specifications based on the One_Max_Length
def toolboxOneMaxLength(toolbox, ONE_MAX_LENGTH):
    toolbox.register("individualCreator", tools.initRepeat, creator.Individual, toolbox.zeroOrOne, ONE_MAX_LENGTH)
    toolbox.register("mutate", tools.mutFlipBit, indpb=1.0/ONE_MAX_LENGTH)

# Changing Toolbox roulette wheel based on 'choice' value
def toolboxRouletteWheel(toolbox,choice):
    if choice == 'selRoulette':
        toolbox.register("select", tools.selRoulette)
    elif choice == 'selTournament':
        toolbox.register("select", tools.selTournament, tournsize=3)

def printVals(showGen=True):
    isTrue = False
    i = 1
    while isTrue == False:
        isTrue, pop, gen = main(i, False, 1)
        i+=1
    print('One_Max_Length:', ONE_MAX_LENGTH)
    print('Smallest Population Size Required:',pop)
    if showGen == True:
        print('Smallest # of Generations Required:',gen)
    print()
    return pop, gen
```

### Problem statements

```python
## Generate all possible tours of the cities 
def exact_TSP(cities):
    # and choose the shortest one
    return shortest(alltours(cities))

## Return the tour with the minimum total distance
def shortest(tours): 
    return min(tours, key=total_distance)
```

```python
# The permutation function is already defined in 'itertools'
alltours = itertools.permutations 
```

### Representing cities and distance

```python
## Fitness function
# Distance between each pair of consecutive cities
def total_distance(tour):
    # could be used any other measure
    return sum(distance(tour[i], tour[i-1]) 
               for i in range(len(tour)))
```

```python
# Constructor for new cities
# using the class 'complex': i+j
# can be used any other strategy
City = complex 
```

```python
# Distance between two cities
def distance(A, B): 
    return abs(A - B)
```

```python
# lower & upper bounds
Xlow, Xupp, Ylow, Yupp  = 10, 890, 10, 590

# n cities with random coordinates
def generate_cities(n):
    return set(City(random.randrange(Xlow, Xupp), 
                    random.randrange(Ylow, Yupp)) 
               for c in range(n))
```

```python
# set of random cities
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
cities = generate_cities(50)
```

```python
## Return a list of tours, each a permutation of cities, but each one starting with the same city
def all_non_redundant_tours(cities):
    start = first(cities)
    return [[start] + list(tour)
            for tour in itertools.permutations(cities - {start})]


## Start iterating over collection, and return the first element
def first(collection):
    for x in collection: return x
    

## Generate all possible tours of the cities and choose the shortest one
def exact_non_redundant_TSP(cities):
    return shortest(all_non_redundant_tours(cities))
```

```python
# plotting the individuals and their fitness (color-coded)
def plot_population(record, min_fitness, max_fitness):
    # darker individuals have a better fitness
    pop = record['pop']
    fits = record['fitness']
    index = sorted(range(len(fits)), key=lambda k: fits[k])

    norm=colors.Normalize(vmin=min_fitness,
                          vmax=max_fitness)
    sm = cmx.ScalarMappable(norm=norm, 
                            cmap=plt.get_cmap('PuBu'))
    
    for i in range(len(index)):
        color = sm.to_rgba(max_fitness - fits[index[i]][0])
        plot_tour(create_tour(pop[index[i]]), alpha=0.5, color=color)
```

```python
## Plot the tour as blue lines between blue circles, and the starting city as a red square.
def plot_tour(tour, alpha=1, color=None):
    plotline(list(tour) + [tour[0]], alpha=alpha, color=color)
    plotline([tour[0]], style='gD', alpha=alpha, size=10)
    # plt.show()

## blue lines between blue circles
def plotline(points, style='bo-', alpha=1, size=7, color=None):
    # Plot a list of points (complex numbers) in the 2-D plane
    X, Y = XY(points)
    # if defined any color
    if color:
        plt.plot(X, Y, style, alpha=alpha, markersize=size, color=color)
    else:
        plt.plot(X, Y, style, alpha=alpha, markersize=size)

## Given a list of points, return two lists: X coordinates, and Y coordinates
def XY(points):
    # here is the reason to use 'complex' numbers
    # X coordinates = real part, Y coordinates = imaginary part
    return [p.real for p in points], [p.imag for p in points]
```

## Part 4 Comparison with Greedy_TSP()



```python
# Visit the nearest neighbor that is still unvisited
def greedy_TSP(cities):
    start = first(cities)
    tour = [start]
    unvisited = cities - {start}
    while unvisited:
        C = nearest_neighbor(tour[-1], unvisited)
        tour.append(C)
        unvisited.remove(C)
    return tour
```

```python
# Find the city in cities that is nearest to city A
def nearest_neighbor(A, cities):
    return min(cities, key=lambda x: distance(x, A))
```

```python
plot_tour(greedy_TSP(cities))
```

### Pros and cons of GA and Greedy Nearest Neighbor algorithms

Pros of Genetic Algorithm 

•	The search for the optimal solution is conducted on a population of points as opposed to a single point (Electricalvoice, 2017).  
•	An objective function is used i.e., the fitness function in this case which is a better method to find the maximum or the mean compared to derivatives.  
•	Inputted data is subjected to probabilistic transition rules as opposed to deterministic rules so that new outputs are always produced.  
•	It works well to avoid falling into the local minima/maxima trap i.e., it searches for the global optimal solution.  
•	It works well on discrete and continuous data.

Cons of Genetic Algorithm

•	GA can be considered an art. On a large scale, its’ results should be interpreted with caution (Electricalvoice, 2017).   
•	The GA process mainly concentrates on the utilisation of the objective function rather than actual information about the problem at hand. Therefore, finding the correct operators can prove to be difficult.  
•	GA is typically time-consuming and/or highly expensive  


Pros of Greedy Nearest Neighbor

•	It is very easy to implement. This is mainly since it is a non-parametric algorithm (Swift, 2021).  
•	Also, useful to solve NP (nondeterministic polynomial) problems.  
•	Requires less computational effort so it is less time consuming and cheaper.  

Cons of Greedy Nearest Neighbor

•	It may be subject to find the local optimal solution as opposed to the global optimal solution (Swift, 2021).  
•	Understanding and proving the correctness is difficult.   

Many of the pros and cons of each algorithm mentioned here were evident in our solution to solving the TSP. In general, the GA should produce better results as it is a process of natural selection and evolution whereas Greedy Nearest Neighbour never backtracks to find better solutions. Although the Greedy Nearest Neighbour produced better results than the GA, I believe this can be attributed to the difficulty we found in discovering the best operators for the fitness function. This was mentioned as a con to the GA.
